```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\psoni\Desktop\titanic_dataset.csv")
```

```python
print(df.head(2))
#print(df.tail(2))
```

```python
new = df.drop(["name","ticket","cabin","boat","body","home.dest"],axis=1)
print(new.describe(include="all"))
#print(new.describe())
```

```python
new
```

```python
new.dropna(axis=0)
```

```python
print(new.pclass.mode())
new.pclass.fillna(new.pclass.mode()[0])
print(new.head(0))
new.index
```

```python
plt.subplot(3,2,1)
new.pclass.fillna(new.pclass.mode()[0]).plot(kind="kde")
plt.subplot(3,2,2)
new.pclass.fillna(new.pclass.mode()[0]).plot(kind="bar")
plt.subplot(3,2,3)
new.pclass.fillna(new.pclass.mode()[0]).plot(kind="line")
plt.subplot(3,2,4)
new.pclass.fillna(new.pclass.mode()[0]).plot(kind="hist")
plt.subplot(3,2,5)
new.pclass.fillna(new.pclass.mode()[0]).plot(kind="kde")
plt.subplot(3,2,6)
new.pclass.fillna(new.pclass.mode()[0]).plot(kind="box")
plt.show()
```

```python
plt.subplot(3,1,1)
new.fare.plot(kind="hist")
plt.subplot(3,1,2)
new.fare.plot(kind="line")
plt.show()
```

```python
cl = new.pclass
fare = new.fare
plt.subplot(3,1,3)
plt.scatter(cl,fare,s=1.0)
plt.show()
```

```python
grp1 = new.groupby(["pclass","sex"])
#print(grp1.groups)
#print(grp1.get_group((1.0,"female")).describe(include="all"))
print(grp1.get_group((1.0,"female")).count())
```
