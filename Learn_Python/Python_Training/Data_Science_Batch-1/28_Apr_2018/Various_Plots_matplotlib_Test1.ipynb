```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Line plot for random number

```python
y = np.random.random(50)
x = np.arange(50)
plt.plot(x,y)
plt.show()
```

# Scatter plot for random number

```python
y = np.random.random(50)
x = np.arange(50)
plt.scatter(x,y,marker=".",color="g",s=200)
plt.show()
```

# Scatter plot for random number with different point size

```python
y = np.random.random(50)
x = np.arange(50)
for pt in range(len(x)):
    plt.scatter(pt,y[pt],marker=".",color="g",s=200.*np.pi*y[pt]**2)
plt.show()
```

# Scatter plot with connected line for random number with different point size 

```python
y = np.random.random(50)
x = np.arange(50)
for pt in range(len(x)):
    plt.scatter(pt,y[pt],marker=".",color="g",s=200.*np.pi*y[pt]**2)
plt.plot(x,y)
plt.show()
```

# Stem plot for random number

```python
y = np.random.randint(-200,500,20)
x = np.arange(20)
plt.stem(x,y)
plt.show()
```

# Sin wave plot

```python
x = np.linspace(0,50,1000)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

# Bar plot

```python
y = np.random.randint(-200,500,20)
x = np.arange(20)
plt.bar(x,y)
plt.show()
```

# Stacked bar plot

```python
y1 = np.random.randint(-200,500,20)
y2 = np.random.randint(-100,500,20)
x = np.arange(20)
plt.bar(x,y1,label="y1")
plt.bar(x,y2,bottom=y1,label="y2")
plt.legend()
plt.show()
```
