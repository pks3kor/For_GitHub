```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line, Scatter,Bar,barH,Stacked bar,Hist,stem

```python
x = np.arange(101)
y = np.random.randint(1,100,101)
#print x
#print y
```

# Line Plot

```python
plt.plot(x,y)
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.legend()
plt.show()
```

```python
for la in range(len(x)):
    if y[la] >50:
        plt.scatter(la,y[la],color="g")
    else:
        plt.scatter(la,y[la],color="b")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.plot(x,y)
for la in range(len(x)):
    plt.text(la,y[la]+.5,(la,y[la]))
plt.legend()

plt.show()
```

```python
plt.bar(x[:20],y[:20],label="male",color="m")
plt.bar(x[20:40],y[20:40],label="female")
#plt.bar(x[20:40],y[20:40])
for la in range(len(x[:20])):
    plt.text(la,y[la],y[la])
plt.legend()
plt.show()
```

```python
y1 = y[:20]
y2 = y[20:40]
print y1
print y2
for idx in range(len(x[:20])):
    plt.bar(idx,y1[idx])
    plt.bar(idx,y2[idx],bottom=y1[idx])
for la in range(len(x[:20])):
    plt.text(la,y1[la]+2,y1[la])
    plt.text(la,y1[la]+y2[la]+2,y2[la])
plt.legend()
plt.show()
```

```python
y1 = y[:20]
y2 = y[20:40]
print y1
print y2
for idx in range(len(x[:20])):
    plt.bar(idx,y1[idx])
    plt.bar(idx+.5,y2[idx])
for la in range(len(x[:20])):
    plt.text(la,y1[la]+2,y1[la])
    plt.text(la+.5,y2[la],y2[la])
plt.legend()
plt.show()
```
