```python
#print(df.head(1))
df_notPlaced_with_Exp = df.where((df.Experience_Yrs>0) & (df.Placement=="Not Placed")).dropna()
df_Placed_with_Exp = df.where((df.Experience_Yrs>0) & (df.Placement=="Placed")).dropna()
plt.bar("Placed_with_Exp",df_Placed_with_Exp.shape[0],label="Placed_with_Exp")
plt.bar("Not_Placed_with_Exp",df_notPlaced_with_Exp.shape[0],label="Not_Placed_with_Exp")
#print df_notPlaced_with_Exp.shape
plt.legend()
plt.show()


```

```python
#print(df.head(1))
#df_notPlaced_salary = df.where((df.Salary>0) & (df.Placement=="Not Placed")).dropna()
plt.subplot(2,1,1)
df_Placed_salary = df.where((df.Salary>0) & (df.Placement=="Placed")).dropna()
df_Placed_salary.Salary.plot(kind="box")
plt.subplot(2,1,2)
df_Placed_salary.Salary.plot(kind="hist")
```

```python
bottom_25 = df_Placed_salary.where((df_Placed_salary.Salary<=220000.00000)).dropna()
top_25 = df_Placed_salary.where((df_Placed_salary.Salary>=500000.00000)).dropna()
plt.bar("bottom_25",bottom_25.shape[0])
plt.text("bottom_25",bottom_25.shape[0],bottom_25.shape[0])
plt.bar("top_25",top_25.shape[0])
plt.text("top_25",top_25.shape[0],top_25.shape[0])
plt.show()
```
