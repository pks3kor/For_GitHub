## Count of parch by class

```python
cnt = np.array([new.parch[new.parch==x].count() for x in range(1,4)])
print cnt
#lt.bar([1,2,3],cnt,color = ["y","m","b"],label=["Class-1","Class-2","Class-3"])
plt.bar([1],cnt[0],color = ["y"],label=["Class-1"])
plt.bar([2],cnt[1],color = ["m"],label=["Class-2"])
plt.bar([3],cnt[2],color = ["b"],label=["Class-3"])
plt.legend()
plt.show()
```

```python
pclass = {tmp:new.pclass[new.pclass==tmp].count() for tmp in set(new.pclass)}
survived = {tmp:new.survived[new.survived==tmp].count() for tmp in set(new.survived)}
sex = {tmp:new.sex[new.sex==tmp].count() for tmp in set(new.sex)}
sibsp = {tmp:new.sibsp[new.sibsp==tmp].count() for tmp in set(new.sibsp)}
parch = {tmp:new.parch[new.parch==tmp].count() for tmp in set(new.parch)}
embarked = {tmp:new.embarked[new.embarked==tmp].count() for tmp in set(new.embarked)}
```

```python
for k in sex:
    plt.bar(k,sex[k],label=k)
plt.title("sex")
plt.legend()
plt.show()
```

```python
for k in survived:
    #print k
    plt.bar(k,survived[k],label=k)
plt.title("survived")
plt.legend()
plt.show()
```

```python
for k in sibsp:
    #print k
    plt.bar(k,sibsp[k],label=k)
plt.title("sibsp")
plt.legend()
plt.show()
```

```python
for k in parch:
    plt.bar(k,parch[k],label=k)
    if k >=0:
        plt.text(k,parch[k]+4,parch[k])
plt.title("parch")
plt.legend()
plt.show()
```
